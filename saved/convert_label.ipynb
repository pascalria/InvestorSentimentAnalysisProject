```python
import csv
import pandas as pd
```

```python
with open(r'E:\resources\project\bert\bert-master\data_collection\saved\labels.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    label = list(reader)
```

```python
print(label)
```

```python
for i in range(1, len(label)):
    print(label[i][0])
    if label[i][0] == '1':
        label[i][0] = 'negative'
    elif label[i][0] == '2':
        label[i][0] = 'neutral'
    else:
        label[i][0] = 'positive'
```

```python
with open(r'E:\resources\project\bert\bert-master\data_collection\saved\new_labels.csv', "w+", encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    for la in label:
        print(la)
        writer.writerow(la)
```
